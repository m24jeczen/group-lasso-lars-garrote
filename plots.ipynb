```python
import importlib
from experiments import perform_experiment
```

```python
df_avg, df_all = perform_experiment(n_repeats=50)
```

```python

import matplotlib.pyplot as plt

metrics = ["MSE", "TPR", "FPR"]
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
for i, metric in enumerate(metrics):
    axs[i].bar(df_avg["Method"], df_avg[metric], color=["skyblue", "lightgreen", "salmon"])
    axs[i].set_title(metric)
    axs[i].set_ylabel(metric)
    axs[i].set_ylim(0, 1.1 if metric != "MSE" else None)

plt.tight_layout()
plt.show()
```

```python
def plot_boxplots(df):
    plt.figure(figsize=(6, 4))
    df.boxplot(column="MSE", by="Method")
    plt.title("MSE by Method")
    plt.suptitle("")
    plt.xlabel("Method")
    plt.ylabel("MSE")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
plot_boxplots(df_all)
```
